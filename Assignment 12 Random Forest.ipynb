# Exercise 1 Company Data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import tree
```

```python
df_c=pd.read_csv('D:/Megha/Desktop/DS assignments/Assignment 12 Random Forest/Company_Data.csv')
df_c
```

```python
df_c.dtypes
```

```python
sorted(df_c.Sales.unique())
```

```python
df_c['Sales']=pd.cut(df_c.Sales,bins=[0,6,12,18],labels=['Low','Medium','High'],right=True)
df_c
```

```python
label_encoder=preprocessing.LabelEncoder()
df_c['ShelveLoc']=label_encoder.fit_transform(df_c.ShelveLoc)
df_c['Urban']=label_encoder.fit_transform(df_c.Urban)
df_c['Sales']=label_encoder.fit_transform(df_c.Sales)
df_c['US']=label_encoder.fit_transform(df_c.US)
df_c
```

```python
x=df_c.iloc[:,1:10]
y=df_c.iloc[:,0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
x_train=x_train.values
x_train
```

```python
y_train=y_train.values
y_train
```

```python
x_test=x_test.values
x_test
```

```python
y_test=y_test.values
y_test
```

```python
num_trees = 150
max_features = 3
kfold = KFold(n_splits=10, random_state=7,shuffle=True)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x_train, y_train, cv=kfold)
print(results.mean())
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
model_eval=pd.crosstab(y_pred,y_test)
model_eval
```

```python
pd.Series(y_pred,y_test)
```

```python
#model accuracy
np.mean(y_pred==y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

kfold = KFold(n_splits=10, random_state=7,shuffle=True)
cart=DecisionTreeClassifier()
model2 = BaggingClassifier(base_estimator=cart,n_estimators=num_trees, random_state=7)
results = cross_val_score(model2, x_train, y_train, cv=kfold)
print(results.mean())
```

```python
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
y_pred
```

```python
model_eval2=pd.crosstab(y_pred,y_test)
model_eval2
```

```python
np.mean(y_pred==y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier

kfold = KFold(n_splits=10, random_state=7,shuffle=True)
model3 = AdaBoostClassifier(n_estimators=num_trees, random_state=7)
results = cross_val_score(model3, x_train, y_train, cv=kfold)
print(results.mean())
```

```python
model3.fit(x_train,y_train)
```

```python
y_pred=model3.predict(x_test)
y_pred
```

```python
model_eval3=pd.crosstab(y_pred,y_test)
model_eval3
```

```python
np.mean(y_pred==y_test)
```

```python
df_f['Taxable.Income']=pd.cut(df_f['Taxable.Income'],bins=[10000,30000,100000],labels=['Risky','Good'],right=True)
```

```python
df_f
```

```python
df_f['Taxable.Income']=label_encoder.fit_transform(df_f['Taxable.Income'])
df_f['Undergrad']=label_encoder.fit_transform(df_f['Undergrad'])
df_f['Marital.Status']=label_encoder.fit_transform(df_f['Marital.Status'])
df_f['Urban']=label_encoder.fit_transform(df_f['Urban'])
df_f
```

```python
x=df_f.drop('Taxable.Income',axis=1)
```

```python
y=df_f['Taxable.Income']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train=x_train.values
y_train=y_train.values
x_test=x_test.values
y_test=y_test.values
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model_eval=pd.crosstab(y_pred,y_test)
```

```python
model_eval
```
